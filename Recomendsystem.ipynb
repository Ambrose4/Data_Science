```python
import pandas
movie = pandas.read_csv('Movie_Id_Titles.csv')
movie.head(5)
```

```python
details = pandas.read_csv('file.csv')
details
```

```python
# merge the  two
data = pandas.merge(details, movie, on = 'item_id')
data.head(5)
```

```python
# which movie has the highest rating on average
data.groupby('title')['rating'].mean().sort_values(ascending=False).head(5)
```

```python
# lets see how many people rated
data.groupby('title')['rating'].count().sort_values(ascending=False).head(5)
```

```python
# get movie ratings
ratings = pandas.DataFrame(data.groupby('title')['rating'].mean().sort_values(ascending=False))
# add a column for no of ratings
ratings['no_of_ratings'] = data.groupby('title')['rating'].count()
ratings.head(5)
# we will use this dataframe 'ratings' later
```

```python
# pivot table-it shows which movie the user rated
pivot = data.pivot_table(index='user_id',columns='title', values='rating').head(50)#index is the 1 column
pivot
# nan mean user did not rate movie
```

```python
# select one from the pivot
selected = pivot['Around the World in 80 Days (1956)']

# correlation
# positive means one variable affects the other-one goes up,the other goes up
# negative means one variable affects the oher-one goes up,the other goes down
# 0 correlation-no correlation
# read on correlation

# get similar movies

similar =pivot.corrwith(selected)
similar_movies = pandas.DataFrame(similar, columns =['correlatiions'])
similar_movies
# positive corr-0-1: above o.5-1 0 strong correlation
# show only movies with a correlation above 0.5
# how many people rated


similar_movies = similar_movies.join(ratings['no_of_ratings'])
similar_movies

# sow only the ones with over 200 ratings
similar_movies[similar_movies ['no_of_ratings'] > 300].sort_values('correlatiions', ascending=False).head(20)

# Above we consinder all movies  that had above  50 ratings
# above recomender is using ratings
```
