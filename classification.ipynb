```python
# machine learning is training computer with data
# they are three main areas in ml
# classification,regression,clustering
# skl-check notes on this site 
# classification-indentifying which category something belongs to
import pandas
data = pandas.read_csv('data.csv')
data

# flour-grams,milk-ml sugar-grams,butter-grams
# egg-count,baking powder = spoons,vanilla-spoons,salt-0-10
```

```python
# step 1:split data columns into 2 portions
array = data.values
X = array[:, 0:8] #These are the ingraidients, 8 not counted here ,-8
Y = array[:, 8]

# above split data to x-representing ingraidients and y- representing cake type
```

```python
# Step 2 split rows into two portion
from sklearn import *
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y,
test_size = 0.3, random_state=42)

```

```python
# we now fit the data to the model
# we will use the bayes theorem
# check models on ml models

from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier



model = GradientBoostingClassifier()
model.fit(X_train, Y_train)
# the model learns through the 70% data
# the model gets to know what level of ingradients make a cupcake , a muffin or even an even
```

```python
# gave the model x_test to give its prediction

predictions =  model.predict(X_test)
predictions, Y_test
from sklearn.metrics import accuracy_score
print("Accuracy", accuracy_score(Y_test, predictions))


# below we notice that the predicted 14 of them are right,failed one
# that is 95% pass
```

```python
# lets now create a recipe to bake a cake
recipe = [[58, 20, 15, 84, 9, 20, 1, 0]]
outcome = model.predict(recipe)
print('you will be preparing a', outcome)

# the model does not give a right guess,its an informed guess,its a gues from learning
```
