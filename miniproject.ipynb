```python

import pandas
data = pandas.read_csv('realestate.csv')
data.tail(15)
```

```python
array = data.values
X = array[:,1:4] #bedrooms, bathrooms, size
Y = array[:,4] #price

```

```python

from sklearn import *
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y,
test_size = 0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
predictions, Y_test
```

```python
from sklearn.metrics import r2_score
print('Accuracy',r2_score (predictions, Y_test))
```

```python
bedroom = int(input("How many Bedrooms"))
bathroom = int(input("How many Bathrooms"))
size = int(input("which Size"))

expense = [[bedroom, bathroom, size]] #the person budget
outcome = model.predict(expense)
print('you will buy', outcome)
```
