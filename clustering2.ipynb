```python
import pandas
data = pandas.read_csv('Mall_Customers.csv')
data.tail(50)
# Age in years, income thousand usd per year,score -out of 100
# The mall wants diffrent group with similarities
# The mall would use that for giving offers to its customers
# cluster you dont train or test
```

```python
# check any empties AND  remove
data.dropna(inplace=True)# remove the
data.isnull().sum()
```

```python
array = data.values
X = array[:,2:5] #age ,income ,score
# clustering is unsupervised, no y variables,hence no testing
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=14, random_state = 42)
model.fit(X)
```

```python
# LETS THE CLUSTERS THAT Kmeans CREATED
means =model.cluster_centers_
clusters = pandas.DataFrame(means, columns=['Age', 'Income', 'Score' ])
clusters
```

```python
# lets see who is in cluster,every record joins a given cluster
data['no'] = model.labels_
data[data['no'] == 13].to_csv('clueter13.csv')
```
