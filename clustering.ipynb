```python

import pandas
data = pandas.read_csv('AirlinesCluster.csv')
data
```

```python
data.isnull().sum()
```

```python
subset = data[['FlightMiles','FlightTrans', 'DaysSinceEnroll']]
subset
```

```python
array = subset.values
X = array[:, 0:3]
X
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=5)
model.fit(X)
```

```python
centronoids = model.cluster_centers_
centronoids
```

```python
clusters = pandas.DataFrame(centronoids, columns = ['FlightMiles', 'FlightsTrans', 'DaysSinceEnroll'])
clusters
```
